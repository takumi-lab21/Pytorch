## 自動微分

```python
import torch

x = torch.ones(5)  # input tensor
y = torch.zeros(3)  # expected output
w = torch.randn(5, 3, requires_grad=True)
b = torch.randn(3, requires_grad=True)
z = torch.matmul(x, w)+b
loss = torch.nn.functional.binary_cross_entropy_with_logits(z, y)

print('Gradient function for z =',z.grad_fn)
print('Gradient function for loss =', loss.grad_fn)


loss.backward()
print("勾配の値: ")
print(w.grad)
print(b.grad)
```

勾配:<br>
- $\frac{\partial loss}{\partial w}$ 、$\frac{\partial loss}{\partial b}$ のこと
- 要するに、パラメータをどのように動かせば、lossが下がるか？という情報

勾配が計算できる条件:<br>
- 計算グラフのleaf nodeである
- require_gradが Trueの場合


<img src="https://pytorch.org/tutorials/_images/comp-graph.png" width=700px alt="海の写真" title="空と海">

```python
# 勾配計算のキャンセル

z = torch.matmul(x, w)+b
print(z.requires_grad)

with torch.no_grad():
    z = torch.matmul(x, w)+b
print(z.requires_grad)
```
